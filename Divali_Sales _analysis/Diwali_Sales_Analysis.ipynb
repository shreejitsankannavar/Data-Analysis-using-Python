```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df['Amount'].fillna(df['Amount'].mean(),inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

# Exploratory Data Analysis

```python
#gender
ax=sns.countplot(x='Gender',data=df)
for container in ax.containers:
    ax.bar_label(container)
```

```python
gb=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
gb
```

```python
sns.barplot(x='Gender',y='Amount',data=gb)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
ax=sns.countplot(x="Age Group",data=df,hue="Gender")
for container in ax.containers:
    ax.bar_label(container)
```

```python
gb=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=gb)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# state

```python
st=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
plt.figure.figsize=(20,5)
sns.barplot(x='State',y='Orders',data=st)
```

```python
st=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(12)
plt.figure.figsize=(20,5)
sns.barplot(x='State',y='Amount',data=st)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital status

```python
ax=sns.countplot(x='Marital_Status',data=df)
plt.figure.figsize=(7,5)
for container in ax.containers:
    ax.bar_label(container)
```

```python
gb=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=gb,hue='Gender')
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,5)})
for container in ax.containers:
    ax.bar_label(container)
```

```python
gb=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=gb)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product Category

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(25,10)})
for container in ax.containers:
    ax.bar_label(container)
```

```python
gb=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,10)})
sns.barplot(x='Product_Category',y='Amount',data=gb)
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
gb=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
sns.set(rc={'figure.figsize':(35,10)})
sns.barplot(x='Product_ID',y='Orders',data=gb)
```

# Conclusion:

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
